#### Популяція та вибірка

#### Справедлива (репрезентативна) vs Несправедлива вибірка

##### Що таке справедлива вибірка?

Вона має збігатися з популяцією за:
- пропорціями сегментів,
- розподілом поведінки,
- розподілом платформ,
- країнами,
- активністю,
- часом.

---

####  Приклад: справедлива vs несправедлива вибірка

# Типи даних та Основи описової статистики 

---

# 1. Типи даних у статистиці

Статистика працює з різними типами даних.  
Тип даних визначає:
- які операції дозволені,
- які метрики можна рахувати,
- які графіки можна будувати,
- які статистичні тести застосовувати.

Типи даних поділяються на три групи:

1. **За природою значень**
2. **За рівнем вимірювання (Nominal / Ordinal / Interval / Ratio)**
3. **За поведінкою (discrete / continuous)**

---

# 1.1. Типи даних за природою значень

##  Категоріальні (Categorical)
Дані, які описують *категорію*, а не число.

Приклади:
- країна (UA, PL, KZ)
- пристрій (Android, iOS)
- жанр фільму
- тарифний план

Реальні приклади (4):
1. **Стримінг:** жанр контенту (драма, комедія) → не можна усереднювати.
2. **Підтримка:** тип звернення (платіж, технічна проблема).
3. **Маркетинг:** канал трафіку (Facebook, Google Ads).
4. **E-commerce:** статус замовлення (paid, canceled, returned).

---

##  Числові (Numeric)
Дані, що мають числове значення.

### Поділ:
- дискретні (штучні ← рахуються)
- неперервні (безперервні ← вимірюються)

---

##  Дискретні (Discrete)
Можуть приймати *лише цілі значення*.

Приклади:
- кількість переглядів
- кількість покупок
- кількість повідомлень у чаті

4 реальних приклади:
1. **Стримінг:** скільки разів користувач натиснув pause.
2. **Маркетинг:** кількість конверсій.
3. **Підтримка:** кількість повідомлень у чаті.
4. **E-commerce:** кількість товарів у кошику.

---

##  Неперервні (Continuous)
Можуть приймати *будь-яке значення на шкалі*.

Приклади:
- тривалість перегляду (секунди, хвилини),
- сума платежу,
- затримка сервера (ms),
- час відповіді агента.

4 реальних приклади:
1. **Стримінг:** тривалість сесії перегляду.
2. **Підтримка:** час від першого повідомлення до відповіді.
3. **Маркетинг:** середня ціна за клік (CPC).
4. **E-commerce:** вага товару.

---

# 1.2. Типи даних за *рівнем вимірювання*

##  Nominal (номінальні)
Категорії без порядку:
- країна,
- пристрій,
- жанр.

Не можна:
- порівнювати «більше/менше»,
- рахувати середнє.

Можна:
- частоти,
- моду.

4 приклади:
- UA vs KZ vs PL (країна перегляду)
- Android / iOS (платформа)
- жанр фільму
- статус замовлення

---

##  Ordinal (порядкові)
Є порядок, але немає відстані.

Приклади:
- рейтинг відгуку ★☆☆☆☆–★★★★★,
- пріоритет звернення (низький, середній, високий),
- рівень підписки (Базова → Стандарт → Преміум).

Реальні приклади:
1. **Підтримка:** рівень задоволеності (1–5).
2. **Стримінг:** «подобається / нейтрально / не подобається».
3. **Маркетинг:** “низька / середня / висока ймовірність конверсії”.
4. **E-commerce:** “маленький / середній / великий розмір”.

---

##  Interval (інтервальні)
Відстань між значеннями відома, але немає абсолютного нуля.

Приклад:
- дата (2023, 2024)
- температура (°C)

---

## Ratio (співвідношення)
Є:
- порядок,
- відстань,
- абсолютний нуль.

Приклади:
- сума транзакції,
- кількість переглядів,
- час перегляду,
- швидкість завантаження.

У 90% продуктів дані — Ratio.

### Дисперсія

---

####  Що таке дисперсія 

> **Дисперсія показує, наскільки “розкидані” значення навколо середнього.**

- Якщо всі значення близькі одне до одного → дисперсія **мала**  
- Якщо значення сильно скачуть → дисперсія **велика**

І це напряму відповідає на питання:
- “чи стабільний час відповіді підтримки?”
- “чи стабільний watch time?”
- “чи стабільний CPC?”
- “чи стабільний чек?”

---


### Формула:

[
Var(x) = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2
]

---

#### Формула

#####  Крок 1. (x_i - \bar{x})

**Що робимо?**
Дивимося, наскільки кожне значення відрізняється від середнього.

**Логмув:**
Як далеко кожен користувач “пішов” від типового значення?

**Приклад (watch time):**
Користувачі: `10, 12, 100`

Середнє = 40
Відхилення:

* 10 – 40 = –30
* 12 – 40 = –28
* 100 – 40 = 60

####  Проблема

Якщо просто усереднити відхилення, вони скасують одне одного.

10 нижче середнього → –30
100 вище середнього → +60

→ У сумі буде може бути 9.

---

#### Крок 2. ((x_i - \bar{x})^2)

**Навіщо квадрат?**

1. Щоб не було від’ємних розкидів
2. Щоб “покарати” великі відхилення
3. Щоб зробити нестабільність **видимою**

**Приклад:**

| Значення | Відхилення | Квадрат |
| -------- | ---------- | ------- |
| 10       | –30        | 900     |
| 12       | –28        | 784     |
| 100      | +60        | 3600    |

 **Важливий факт:**
Квадрат робить ВСЕ аналізоване більш чесним.
Без нього 60 хвилин і 20 хвилин були б “майже однакові”.

А квадрат робить:

* 20² = 400
* 60² = 3600

→ **Велике відхилення стає реально більшим.**

---

#### Крок 3. Сума квадратів

[
\sum (x_i - \bar{x})^2
]

Збираємо вcе.

---

#### Крок 4. Ділення на n

[
\frac{1}{n}
]

Це означає:

> “в середньому розкид по всій вибірці”.

---

Якщо дані в:

* хвилинах → дисперсія в хвилинах²
* гривнях → дисперсія в гривнях²
* метрах → метрах²

Це НЕ інтерпретується напряму.
Тому дисперсія не використовується напряму для висновків — тільки в формулах.

---

#### 2. СТАНДАРТНЕ ВІДХИЛЕННЯ (STD)

##### ✔ Формула:

[
Std = \sqrt{Var}
]

---

##### КОРІНЬ

Щоб **повернути одиниці вимірювання назад**.

Було:

* хвилини²
* гривні²

Беремо sqrt:

→ хвилини
→ гривні

Std — інтерпретована величина.


---



####  Приклад

**Data:**

```
[5, 6, 7, 8, 400]
```

* mean = 85.2
* var ≈ 15278
* std ≈ 123.6
* CV = 1.45 →  розкид

### Інтерпретація:

* Середнє **бреше** \
* Типова поведінка — 5–8 хв \
* 400 хв = автоплей/фоновий перегляд

**Медіана = правильна метрика** \

Як зрозуміти “мала” чи “велика” дисперсія?

CV (коефіцієнт варіації) - std / mean * 100

#### Правило 1 — CV:

| CV       | Що означає            | Дії                          |
| -------- | --------------------- | ---------------------------- |
| <0.25    | стабільні дані        | mean можна                   |
| 0.25–0.5 | слабка нестабільність | mean можна, але median теж   |
| 0.5–1.0  | сильна нестабільність | median + IQR                 |
| >1.0     | хаос                  | mean заборонено, сегментація |

---

### Якщо дисперсія мала:

* дані чисті
* mean репрезентує
* можна будувати A/B
* можна будувати тренди
* можна будувати регресію

#### Якщо дисперсія велика:

* mean бреше
* median + IQR — правильні метрики
* потрібен boxplot
* потрібна сегментація
* потрібен пошук вибросів
* агрегувати дані НЕ можна (все змажеться)

---

#### Практичний чеклист 

1️⃣ Порахуй mean, std, CV \
2️⃣ Якщо CV > 0.5 → переходь на median + IQR \
3️⃣ Побудуй boxplot \
4️⃣ Перевір верхні хвости (p90, p95, p99) \
5️⃣ Подивись сегменти \
6️⃣ Подивись, чи не нормальний розподіл \
7️⃣ Якщо heavy-tail — mean не використовувати


## 1. IQR (Interquartile Range) — Міжквартильний розмах

IQR — це показник «чистого», стабільного розкиду даних.
Він описує **середні 50% значень**, і не залежить від викидів.

---

## 1.1. Формула

p = 0.25 * (n - 1) + 1 -> 1.25

Q1 = 25-й перцентиль  
Q3 = 75-й перцентиль  

\[
IQR = Q3 - Q1
\]

---

## 1.2. Навіщо потрібно

IQR відповідає на питання:

> *“Який типічний розкид у 50% найпоширеніших значень?”*

На відміну від std, IQR НЕ реагує на великі викиди.
Тому саме IQR — ключ до реальних бізнес-даних.

---

## 1.3. приклад 

Дані:
    [3, 4, 5, 6, 7, 8, 100]


Q1 = 4  
Q3 = 8  
IQR = 8 – 4 = **4**

→ Хоча 100 — величезний виброс, IQR все одно стабільний.

---

## 1.4. Навіщо потрібен IQR?

### IQR — ключовий інструмент у 4 областях:

###  1) Виявлення вибросів  
Правило Тьюкі:

\[
Lower = Q1 - 1.5 \cdot IQR
\]
\[
Upper = Q3 + 1.5 \cdot IQR
\]

---

###  2) Аналіз перекошених (skewed) даних  
Watch time, час відповіді, CTR, CPC — завжди НЕ нормальні.

IQR дає реальний розкид.





````python
# Середнє (Mean). Середнє(AVG) = сума всіх значень / кількість значень

def mean(arr):
    total = 0
    count = 0
    for x in arr:
        total += x
        count += 1
    return total / count

print(mean([10, 20, 30, 40]))
````

````python
# Медіана (Median). Медіана — центральне значення після сортування
# Медіана не реагує на викиди → максимально стійке середнє
def median(arr):
    arr_sorted = sorted(arr)
    n = len(arr_sorted)
    
    if n % 2 == 1:
        return arr_sorted[n // 2]
    else:
        mid1 = arr_sorted[n//2 - 1]
        mid2 = arr_sorted[n//2]
        return (mid1 + mid2) / 2

print(median([10, 50, 2, 7, 99, 1]))
````

````python
# Мода (Mode). Мода — найчастіше значення
def mode(arr): # or using collections
    freq = {}
    for x in arr:
        if x not in freq:
            freq[x] = 0
        freq[x] += 1
    
    max_count = 0
    mode_value = None

    for key in freq:
        if freq[key] > max_count:
            max_count = freq[key]
            mode_value = key
    
    return mode_value

print(mode([2, 3, 3, 5, 3, 10, 5]))
````

````python
# Розмах (Range). Розмах = max − min
def data_range(arr):
    min_v = arr[0]
    max_v = arr[0]

    for x in arr:
        if x < min_v:
            min_v = x
        if x > max_v:
            max_v = x

    return max_v - min_v

print(data_range([3, 10, 20, 6, 50]))

````

````python
# Дисперсія. Дисперсія - середній квадрат відхилень від середнього
def variance(arr):
    m = mean(arr) # or use numpy
    total = 0
    count = 0
    for x in arr:
        total += (x - m) ** 2
        count += 1
    return total / count

print(variance([10, 12, 9, 11, 30]))

````

````python
# Стандартне відхилення (STD). Корінь із variance

def std(var):
    return var ** 0.5 # or using math lib
# CV (Coefficient of Variation)
# CV = STD / Mean
# Порівнює стабільність різних метрик
def cv(arr):
    return std(arr) / mean(arr)

print(cv([10, 12, 9, 11, 30]))
````

````python
# Квартилі, IQR 
# Теорія
# Q1 — 25% значень
# Q3 — 75%
# IQR = Q3 − Q1

# Викиди:
# нижня межа = Q1 − 1.5 * IQR
# верхня межа = Q3 + 1.5 * IQR
def percentile(arr, p):
    arr_sorted = sorted(arr)
    k = (len(arr_sorted)-1) * p
    f = int(k)
    c = f + 1
    if c >= len(arr_sorted):
        return arr_sorted[f]
    return arr_sorted[f] * (c-k) + arr_sorted[c] * (k-f)

def iqr(arr):
    q1 = percentile(arr, 0.25)
    q3 = percentile(arr, 0.75)
    return q3 - q1, q1, q3

print(iqr([1,2,5,7,9,10,12,15,20]))
````
