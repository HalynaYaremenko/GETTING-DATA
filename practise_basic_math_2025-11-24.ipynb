**Задача 1** \
Дані — перегляд (хв): \
[3, 4, 5, 6, 7]

Знайти медіану.

```python
arr = [4, 5, 3, 7, 6]

arr.sort() # sorted 

n = len(arr)
if n % 2 == 1:
    median_ = arr[n // 2]
else:
    median_ = (arr[n // 2 - 1] + arr[n // 2]) / 2

print(median_)
```

**Задача 2** \
Дані: [3, 3, 4, 100] \
Знайти median та mean. Що про них можна сказати?

```python
arr = [3, 3, 4, 100]

arr.sort() # sorted 

n = len(arr)

if n % 2 == 1:
    median_ = arr[n // 2]
else:
    median_ = (arr[n // 2 - 1] + arr[n // 2]) / 2

print("Median", median_)

avg = sum(arr) / n
print("Mean", avg)
```

Дані днів користування сервісом: [1, 2, 2, 2, 3, 90]

```python
arr = [1, 2, 2, 2, 3, 90]

arr.sort() # sorted 

n = len(arr)

if n % 2 == 1:
    median_ = arr[n // 2]
else:
    median_ = (arr[n // 2 - 1] + arr[n // 2]) / 2

print("Median", median_)

avg = sum(arr) / n
print("Mean", avg)

arr_s = [2, 2, 2, 3]
print("Mean", sum(arr_s) / len(arr_s))
```

# Задача 1
Дані: [2, 4, 6, 8]

Потрібно:
- знайти середнє
- знайти квадрати відхилень
- знайти дисперсію
- інтерпретувати результат


```python
arr4 = [2, 4, 6, 8]

avg_arr = sum(arr4) / len(arr4) # знайти середнє

sum_arr = []
for i in arr4:
    sum_arr.append((i - avg_arr) ** 2) # знайти квадрати відхилень

disp = sum(sum_arr) / len(arr4) # знайти дисперсію
disp
```

Задача 2 — два набори з однаковим середнім, але різним розкидом \
A: [4, 5, 6] \
B: [1, 5, 9] 

Середні однакові.  
Який набір стабільніший?

```python
arr_a = [4, 5, 6]
arr_b = [1, 5, 9]


avg_a = sum(arr_a) / len(arr_a)
avg_b = sum(arr_b) / len(arr_b)

sum_arr_a = []
for i in arr_a:
    sum_arr_a.append((i - avg_a) ** 2)

sum_arr_b = []
for j in arr_b:
    sum_arr_b.append((j - avg_b) ** 2)

disp_a = sum(sum_arr_a) / len(arr_a)
disp_b = sum(sum_arr_b) / len(arr_b)
print(disp_a, disp_b )
```

### Std 
Дані: [4, 5, 6, 9, 10]  

Порахуй variance і std. 


```python
arr5 = [4, 5, 6, 9, 10]  

avg_arr5 = sum(arr5) / len(arr5) # знайти середнє

sum_arr5 = []
for i in arr5:
    sum_arr5.append((i - avg_arr5) ** 2) # знайти квадрати відхилень

disp_arr = sum(sum_arr5) / len(arr5) # знайти дисперсію

std = disp_arr ** 0.5
print("std", std)

variance = std / avg_arr5
print("variance", variance) # 0,25 - 0,5 слабка нестабільність
```
